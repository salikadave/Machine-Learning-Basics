```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
```

```python
#converting the csv file to a panda dataframe
# 4 possible outputs:- unacc, acc, good, vgood
dataFrame = pd.read_csv("CarDataset.csv")
dataFrame.head()
```

```python
#assign labels
df_Label = dataFrame['acceptability']
print(df_Label.head())
```

```python
#assign features
df_Features = dataFrame.drop('acceptability',1)

#replacing vhigh-4, high-3, med-2, low-1 in 'buying'
df_Features['buying'].replace("vhigh",4, inplace=True)
df_Features['buying'].replace("high",3, inplace=True)
df_Features['buying'].replace("med",2, inplace=True)
df_Features['buying'].replace("low",1, inplace=True)

#replacing vhigh-4, high-3, med-2, low-1 in 'maint_cost'
df_Features['maint_cost'].replace("vhigh",4, inplace=True)
df_Features['maint_cost'].replace("high",3, inplace=True)
df_Features['maint_cost'].replace("med",2, inplace=True)
df_Features['maint_cost'].replace("low",1, inplace=True)

#replacing 5more by 5 in 'doors'
df_Features.replace("5more",5, inplace=True)

#replacing more by 5 in 'persons'
df_Features.replace("more",5, inplace=True)

#replacing small-1, med-2, big-3 in 'lug_boot' 
df_Features['lug_boot'].replace("small",1, inplace=True)
df_Features['lug_boot'].replace("med",2, inplace=True)
df_Features['lug_boot'].replace("big",3, inplace=True)

#replacing low-1, med-2, high-3 in 'safety'
df_Features['safety'].replace("low",1, inplace=True)
df_Features['safety'].replace("med",2, inplace=True)
df_Features['safety'].replace("high",3, inplace=True)

df_Features.head(100)
```

```python
#assigning binary codes to the predicted outputs
label = []
for lab in df_Label:
    if lab == "unacc":
        label.append([1,0,0,0])
    elif lab == "acc":
        label.append([0,1,0,0])
    elif lab == "good":
        label.append([0,0,1,0])
    else:
        label.append([0,0,0,1])
#print(label)
```

```python
#converting the label and features to array datatypes
dataArr = np.array(df_Features)
#print(dataArr)
labelArr = np.array(label)
#print(labelArr)
#print(dataArr.shape,labelArr.shape)
```

```python
#splitting the data for testing and training
#x - features (here - array containing 6 attribute values of the car)
#y - labels (here - array containing 4 possible outcomes)
x_train, x_test, y_train, y_test = train_test_split(dataArr, labelArr, test_size=0.2, random_state = 42)
x_train.shape
x_test.shape
```

```python
#creating the neural network
model = Sequential()
model.add(Dense(500, input_dim=6, activation='sigmoid'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(4, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
model.fit(x_train,y_train, epochs=1000, batch_size=70, validation_data=(x_test,y_test))
```

```python
#testing the model

test1 = np.array([[4,4,2,2,1,1]])  #belongs to "unacceptable"
test2 = np.array([[4,1,5,4,1,3]])  #belongs to "acceptable"
test3 = np.array([[2,1,2,4,1,3]])  #belongs to "good"
test4 = np.array([[2,2,3,5,2,3]])  #belongs to "vgood"
result = model.predict_classes(test4)
print(result)
if result==0:
    print("Unacceptable")
elif result==1:
    print("Acceptable")
elif result==2:
    print("Good")
else:
    print("Very Good")
```
